```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
salary_train = pd.read_csv("C:\\My Computer\\Documents\\Excler\\Assignment\\Naive Bayes\\SalaryData_Train.csv")
salary_test = pd.read_csv("C:\\My Computer\\Documents\\Excler\\Assignment\\Naive Bayes\\SalaryData_Test.csv")
```

```python
category = ['workclass', 'education','maritalstatus','occupation','relationship','race','sex','native']
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
for i in category:
    salary_train[i] = labelencoder.fit_transform(salary_train[i])
    salary_test[i] = labelencoder.fit_transform(salary_test[i])
```

```python
salary_train.head()
```

```python
X_train = salary_train.iloc[:,0:13]
X_test = salary_test.iloc[:,0:13]
y_train = salary_train.iloc[:,13]
y_test = salary_test.iloc[:,13]
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
X_train = norm_func(X_train)
X_test =  norm_func(X_test)
```

```python
X_train
```

### Naive Bayes Model

```python
from sklearn.naive_bayes import MultinomialNB as MB
from sklearn.naive_bayes import GaussianNB as GB

# Multinomial Naive Bayes
classifier_mb = MB()
classifier_mb.fit(X_train,y_train)
train_pred_m = classifier_mb.predict(X_train)
accuracy_train_m = np.mean(train_pred_m==y_train) 

test_pred_m = classifier_mb.predict(X_test)
accuracy_test_m = np.mean(test_pred_m==y_test)
```

```python
print(accuracy_test_m, accuracy_train_m)
```

```python
# Gaussian Naive Bayes
classifier_gb = GB()
classifier_gb.fit(X_train,y_train)
train_pred_g = classifier_gb.predict(X_train)
accuracy_train_g = np.mean(train_pred_g==y_train) 

test_pred_g = classifier_gb.predict(X_test)
accuracy_test_g = np.mean(test_pred_g==y_test)
```

```python
print(accuracy_test_g, accuracy_train_g)
```
